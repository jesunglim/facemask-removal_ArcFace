```python
import matplotlib.pyplot as plt

import os
import shutil
```

```python
og_path = '/home/ielab/project/samples/OG_loss.txt'
mo_path = '/home/ielab/project/samples/mosaic_loss.txt'
msk_path = '/home/ielab/project/samples/masked_loss.txt'
gan_path = '/home/ielab/project/samples/gan_loss.txt'
```

```python
og_f = open(og_path, 'r')
mo_f = open(mo_path, 'r')
msk_f = open(msk_path, 'r')
gan_f = open(gan_path, 'r')
```

```python
og_lines = og_f.readlines()
mo_lines = mo_f.readlines()
msk_lines = msk_f.readlines()
gan_lines = gan_f.readlines()
```

```python
og_lines = [line.rstrip() for line in og_lines]
mo_lines = [line.rstrip() for line in mo_lines]
msk_lines = [line.rstrip() for line in msk_lines]
gan_lines = [line.rstrip() for line in gan_lines]

og_lines = list(map(float, og_lines))
mo_lines = list(map(float, mo_lines))
msk_lines = list(map(float, msk_lines))
gan_lines = list(map(float, gan_lines))
```

```python
len(gan_lines)
```

```python
import numpy as np
```

```python
plt.subplots(constrained_layout=True)

plt.subplot(2, 2, 1)
plt.plot(og_lines)
plt.title('Original')


plt.subplot(2, 2, 2)
plt.plot(mo_lines)
plt.title('Mosaic')


plt.subplot(2, 2, 3)
plt.plot(msk_lines)
plt.title('Masked')


plt.subplot(2, 2, 4)
plt.plot(gan_lines)
plt.title('GAN')



plt.show()
```

```python
num = 100

plt.subplots(constrained_layout=True)

plt.subplot(2, 2, 1)
plt.plot(og_lines[:num])
plt.title('Original')


plt.subplot(2, 2, 2)
plt.plot(mo_lines[:num])
plt.title('Mosaic')


plt.subplot(2, 2, 3)
plt.plot(msk_lines[:num])
plt.title('Masked')


plt.subplot(2, 2, 4)
plt.plot(gan_lines[:num])
plt.title('GAN')



plt.show()
```
